```python
new_df = df.drop(df.index[[76, 70]], axis = 0)
```

```python
#preparing model
ml_new = smf.ols('MPG~WT+VOL+HP+SP', data=new_df).fit()
```

```python
ml_new.params
```

```python
ml_new.summary()
```

```python
print(ml_new.conf_int(0.01))
```

```python
#predict values of MPG
mpg_pred = ml_new.predict(new_df[['WT', 'VOL', 'HP', 'SP']])
mpg_pred
```

```python
new_df.head()
```

```python
#calculating vif values of independent variables.
rsq_hp = smf.ols('HP~WT+VOL+SP', data= new_df).fit().rsquared
vif_hp = 1/(1-rsq_hp)
vif_hp
```

```python
rsq_wt = smf.ols('WT~HP+VOL+SP', data=new_df).fit().rsquared
vif_wt = 1/(1-rsq_wt)
vif_wt
```

```python
rsq_vol = smf.ols('VOL~HP+WT+SP', data=new_df).fit().rsquared
vif_vol = 1/(1-rsq_vol)
vif_vol
```

```python
rsq_sp = smf.ols('SP~HP+WT+VOL', data = new_df).fit().rsquared
vif_sp = 1/(1-rsq_sp)
vif_sp
```

```python
#storing vif values in a dataframe
d1 = {'variables': ['HP', 'WT', 'VOL', 'SP'], 'VIF':[vif_hp, vif_wt, vif_vol, vif_sp]}
vif_frame = pd.DataFrame(d1)
vif_frame
```

```python
#final model
final_ml = smf.ols('MPG~HP+SP+VOL', data = new_df).fit()
```

```python
final_ml.params
```

```python
final_ml.summary()
```

```python
mpg_pred = final_ml.predict(new_df)
```

```python
import statsmodels.api as sm
#added variable plot for the final model
sm.graphics.plot_partregress_grid(final_ml)
```

```python
#Linearity
#observed values v/s Fitted Values
plt.scatter(new_df.MPG, mpg_pred, c = 'r')
plt.xlabel('Observed Values')
plt.ylabel('Fitted Values')
```

```python
#Residuals v/s Fitted Values
plt.scatter(mpg_pred, final_ml.resid_pearson, c = 'r')
plt.axhline(y = 0, color = 'blue')
plt.xlabel('fitted_values')
plt.ylabel('residuals')
```

```python
#Normality plot for residuals
#histogram
plt.hist(final_ml.resid_pearson)
```

```python
#QQ plot for residuals
import pylab
import scipy.stats as st
```

```python
#checking residuals are normally distributed
st.probplot(final_ml.resid_pearson, dist = 'norm', plot = pylab)
```

```python
#Homoscedaticity
#Residuals v/s Fitted Values
plt.scatter(mpg_pred, final_ml.resid_pearson, c = 'r')
plt.axhline(y = 0, color = 'blue')
plt.xlabel('fitted_values')
plt.ylabel('residuals')
```

```python
from sklearn.model_selection import train_test_split
cars_train, cars_test = train_test_split(new_df, test_size = 0.2)
```

```python
model_train = smf.ols('MPG~HP+SP+VOL', data = cars_train).fit()
```

```python
train_pred = model_train.predict(cars_train)
train_pred
```

```python
train_resid = train_pred - cars_train.MPG
train_resid
```

```python
train_rmse = np.sqrt(np.mean(train_resid*train_resid))
train_rmse
```

```python
test_pred = model_train.predict(cars_test)
test_pred

```

```python
test_resid = test_pred - cars_test.MPG
test_resid
```

```python
test_rmse = np.sqrt(np.mean(test_resid*test_resid))
test_rmse
```

```python
final = pd.DataFrame({'train_pred': train_pred, 'train_resid':train_resid, 'train_rmse': train_rmse, 'test_pred':test_pred, 'test_resid':test_resid})
final.head(10)
```
